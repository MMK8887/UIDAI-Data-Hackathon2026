### üü¶ CELL 1 ‚Äî Imports & Global Setup

```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

np.random.seed(42)
print("Libraries imported successfully")
```

### üü¶ CELL 2 ‚Äî Path Configuration & Validation

```python
BASE_DATA_DIR = r"C:\Users\HP\Desktop\uidata"

ENROL_PATH = os.path.join(BASE_DATA_DIR, "api_data_aadhar_enrolment")
DEMO_PATH  = os.path.join(BASE_DATA_DIR, "api_data_aadhar_demographic")
BIO_PATH   = os.path.join(BASE_DATA_DIR, "api_data_aadhar_biometric")

print("BASE PATH:", BASE_DATA_DIR)
print("ENROL PATH EXISTS:", os.path.exists(ENROL_PATH))
print("DEMO PATH EXISTS :", os.path.exists(DEMO_PATH))
print("BIO PATH EXISTS  :", os.path.exists(BIO_PATH))
# ‚úÖ Expect all True
```

### üü¶ CELL 3 ‚Äî Loader Function (Verbose)

```python
def load_and_concat(folder_path):
    files = glob.glob(os.path.join(folder_path, "*.csv"))
    print("\n----------------------------------")
    print("Loading from:", folder_path)
    print("CSV files found:", len(files))
    
    if not files:
        raise ValueError("‚ùå No CSV files found")
        
    dfs = []
    for f in files:
        print("‚Üí", os.path.basename(f))
        dfs.append(pd.read_csv(f))
        
    df = pd.concat(dfs, ignore_index=True)
    print("Combined shape:", df.shape)
    return df
```

### üü¶ CELL 4 ‚Äî Standardization Function (Verbose)

```python
def standardize_uidai_df(df, name="DATA"):
    print(f"\n--- Standardizing {name} ---")
    print("Before:", df.shape)
    
    df = df.copy()
    df.columns = df.columns.str.lower().str.strip()
    
    for col in ["state", "district"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()
    
    if "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors="coerce", dayfirst=True)
        df["month"] = df["date"].dt.to_period("M").dt.to_timestamp()
    
    print("After:", df.shape)
    print("Nulls per column:\n", df.isna().sum())
    return df
```

### üü¶ CELL 5 ‚Äî Load Raw Data (FULL INSPECTION)

```python
enrol = standardize_uidai_df(load_and_concat(ENROL_PATH), "ENROL")
demo  = standardize_uidai_df(load_and_concat(DEMO_PATH), "DEMO")
bio   = standardize_uidai_df(load_and_concat(BIO_PATH), "BIO")

print("\n=== DATA LOADED ===")
print("ENROL:", enrol.shape)
print("DEMO :", demo.shape)
print("BIO  :", bio.shape)
```

### üü¶ CELL 6 ‚Äî Raw Schema Inspection

```python
print("\nENROL INFO")
enrol.info()
display(enrol.head())
display(enrol.describe(include="all"))

print("\nDEMO INFO")
demo.info()
display(demo.head())
display(demo.describe(include="all"))

print("\nBIO INFO")
bio.info()
display(bio.head())
display(bio.describe(include="all"))
```

### üü¶ CELL 7 ‚Äî Column Renaming (Verified)

```python
enrol = enrol.rename(columns={
    "age_0_5": "enrol_age_0_5",
    "age_5_17": "enrol_age_5_17",
    "age_18_greater": "enrol_age_18_plus"
})

demo = demo.rename(columns={
    "demo_age_5_17": "demo_age_5_17",
    "demo_age_17_": "demo_age_18_plus"
})

bio = bio.rename(columns={
    "bio_age_5_17": "bio_age_5_17",
    "bio_age_17_": "bio_age_18_plus"
})

for df in [enrol, demo, bio]:
    for col in ["pincode", "source_file"]:
        if col in df.columns:
            df.drop(columns=col, inplace=True)

print("ENROL columns:", enrol.columns.tolist())
print("DEMO columns :", demo.columns.tolist())
print("BIO columns  :", bio.columns.tolist())
```

### üü¶ CELL 8 ‚Äî Monthly Aggregation (INSPECT EVERYTHING)

```python
enrol_m = enrol.groupby(["month", "state", "district"], as_index=False).sum(numeric_only=True)
demo_m  = demo.groupby(["month", "state", "district"], as_index=False).sum(numeric_only=True)
bio_m   = bio.groupby(["month", "state", "district"], as_index=False).sum(numeric_only=True)

print("Aggregated Shapes:")
print("ENROL_M:", enrol_m.shape)
print("DEMO_M :", demo_m.shape)
print("BIO_M  :", bio_m.shape)

display(enrol_m.head())
display(demo_m.head())
display(bio_m.head())
```

### üü¶ CELL 9 ‚Äî Merge Validation

```python
merged = enrol_m.merge(demo_m, on=["month", "state", "district"], how="left")
merged = merged.merge(bio_m, on=["month", "state", "district"], how="left")

print("MERGED SHAPE:", merged.shape)
merged.info()
display(merged.head())
```

### üü¶ CELL 10 ‚Äî Feature Engineering Checks

```python
merged["enrol_total"] = (
    merged["enrol_age_0_5"] +
    merged["enrol_age_5_17"] +
    merged["enrol_age_18_plus"]
)

update_cols = ["demo_age_5_17", "demo_age_18_plus", "bio_age_5_17", "bio_age_18_plus"]
merged[update_cols] = merged[update_cols].fillna(0)

print("ENROL TOTAL STATS")
display(merged["enrol_total"].describe())
```

### üü¶ CELL 11 ‚Äî Rate Computation Inspection

```python
merged["demo_rate_5_17"] = merged["demo_age_5_17"] / merged["enrol_total"]
merged["demo_rate_18_plus"] = merged["demo_age_18_plus"] / merged["enrol_total"]
merged["bio_rate_5_17"] = merged["bio_age_5_17"] / merged["enrol_total"]
merged["bio_rate_18_plus"] = merged["bio_age_18_plus"] / merged["enrol_total"]

rate_cols = [
    "demo_rate_5_17",
    "demo_rate_18_plus",
    "bio_rate_5_17",
    "bio_rate_18_plus"
]

display(merged[rate_cols].describe())
```

### üü¶ CELL 12 ‚Äî Z-Score Inspection

```python
for col in rate_cols:
    mu = merged[col].mean()
    sigma = merged[col].std()
    merged[col + "_z"] = np.where(sigma == 0, 0, (merged[col] - mu) / sigma)

z_cols = [c + "_z" for c in rate_cols]

display(merged[z_cols].describe())
```

### üü¶ CELL 13 ‚Äî Anomaly Detection Diagnostics

```python
merged["stat_anomaly_flag"] = ((merged[z_cols].abs() >= 3).any(axis=1)).astype(int)

X = merged[z_cols].fillna(0).values
iso = IsolationForest(n_estimators=300, contamination=0.05, random_state=42)
merged["ml_anomaly_flag"] = (iso.fit_predict(X) == -1).astype(int)

print("Anomaly Counts:")
print(merged[["stat_anomaly_flag", "ml_anomaly_flag"]].value_counts())
```

### üü¶ CELL 14 ‚Äî DSI Distribution Inspection

```python
merged["base_stress"] = merged[z_cols].abs().mean(axis=1)
merged["DSI"] = merged["base_stress"] * (1 + 0.5 * (merged["stat_anomaly_flag"] + merged["ml_anomaly_flag"]))

display(merged["DSI"].describe())

merged["stress_level"] = pd.qcut(
    merged["DSI"],
    q=[0, 0.7, 0.9, 0.97, 1.0],
    labels=["Low", "Moderate", "High", "Extreme"]
)

print(merged["stress_level"].value_counts())
```

### üü¶ CELL 15 ‚Äî Temporal Change & Early Warning

```python
merged = merged.sort_values(["state", "district", "month"])
merged["DSI_change"] = merged.groupby(["state", "district"])["DSI"].diff()

display(merged["DSI_change"].describe())

threshold = merged["DSI_change"].quantile(0.95)
merged["early_warning"] = (merged["DSI_change"] > threshold).astype(int)

print("Early Warning Count:", merged["early_warning"].sum())
```

### üü¶ CELL 16 ‚Äî Final Results & Plots

```python
latest_month = merged["month"].max()

top_districts = (
    merged[merged["month"] == latest_month]
    .sort_values("DSI", ascending=False)
    .head(10)
)

display(top_districts)

merged.groupby("month")["DSI"].mean().plot(marker="o", figsize=(10,5))
plt.title("Average District Stress Index Over Time")
plt.show()
```
